```python
connection = psycopg2.connect('postgresql://fga@localhost/fga')
```

```python
def query_database(connection, sql, *args, **kwargs):
    if connection.get_transaction_status() == 3:
        connection.reset()
    with connection.cursor() as cursor:
        cursor.execute(sql, *args, **kwargs)
        rows = cursor.fetchall()
        columns = [d[0] for d in cursor.description]
    df = pd.DataFrame(rows, columns=columns)
    return df
```

```python
sql = '''
with events_ as (
    select
        "createdAt" as ts, 
        "currentUser" as user_id,
        "userDevice" as device_id
        
    from events
    
), devices_ as (
    select
        "Id" as device_id,
        "DeviceOS" as client
        
    from devices
    
), results as (
    select
        ts::date as date,
        client,
        count(distinct user_id) as dau
        
    from events_ join devices_ using (device_id)
    
    group by 1, 2
    
)

select * from results

'''

df = query_database(connection, sql)
df.head()
```

```python
sql = ''' 
with events_ as (
    select
        "createdAt" as ts,
        "action" as action,
        "currentUser" as user_id
    
    from events
    
    order by user_id, ts
    
), events_grouped as (
    select
        user_id,
        ts,
        action,
        row_number() over (partition by user_id order by ts) as event_number
        
    from events_
    
), events_step as (
    select
        c.user_id,
        c.ts,
        p.ts as ts_previous,
        c.action,
        p.action as action_previous,
        c.ts::timestamp - p.ts::timestamp as td
        
    from events_grouped p join events_grouped c 
        on p.event_number + 1 = c.event_number 
        and p.user_id = c.user_id
    
), results as (
    select * from events_step
)

select * from results 
    
'''

df = query_database(connection, sql)
df.head()
```

```python
sql = ''' 
with events_ as (
    select
        "createdAt" as ts,
        "action" as action,
        "userDevice" as device_id
    
    from events
    
    order by device_id, ts
    
), events_grouped as (
    select
        device_id,
        ts,
        action,
        row_number() over (partition by device_id order by ts) as event_number
        
    from events_
    
), events_step as (
    select
        c.device_id,
        c.ts,
        p.ts as ts_previous,
        c.action,
        p.action as action_previous,
        c.ts::timestamp - p.ts::timestamp as td
        
    from events_grouped p join events_grouped c 
        on p.event_number + 1 = c.event_number 
        and p.device_id = c.device_id
    
), start_events as (
    select
        device_id,
        ts
        
    from events_
    
    where action = 'APP_COMPONENT_StartApp'
    
), events_first_10_mins as (
    select
        *
        
    from start_events l join events_step r
        on l.device_id = r.device_id
        and r.ts::timestamp - l.ts::timestamp <= interval '10 minute'
        and r.ts::timestamp >= l.ts::timestamp

),  results as (
    select * from events_first_10_mins
)

select * from results
    
'''

df = query_database(connection, sql)
df.head()
```

```python
sql = '''
with events_ as (
    select
        "userDevice" as device_id,
        "createdAt"::timestamp as ts,
        "action" as action
        
    from events
    
), start_events as (
    select
        device_id,
        ts,
        row_number() over (partition by device_id order by ts) as session_id
        
    from events_
    
    where action='APP_COMPONENT_StartApp'
    
), session_events as (
    select
        l.device_id,
        r.ts,
        session_id,
        action
        
    from start_events l join events_ r
        on l.device_id = r.device_id
        and r.ts >= l.ts
        and r.ts < l.ts + interval '10 minute'

), results as (
    select * from session_events
)

select * from results

'''

import datetime
ts_start = datetime.datetime.now()
df = query_database(connection, sql)
ts_end = datetime.datetime.now()
print('duration:', ts_end - ts_start)
df.head()
```

```python
sql = ''' 
with distinct_events as (
    select distinct
        device_id, 
        session_id, 
        action

    from session_events 
    
), event_counts as (
    select
        action,
        "DeviceOS" as client,
        count(1)
        
    from distinct_events l join devices r on l.device_id=r."Id"
    
    group by 1, 2
    
), n_sessions as (
    select
        "DeviceOS" as client,
        count(*)
        
    from (
        select distinct
            device_id,
            session_id
            
        from distinct_events
    ) a join devices on a.device_id=devices."Id"
    
    group by 1
            
), results as (
    select
        client,
        action, 
        l.count as n_actions,
        r.count as n_sessions
        
    from event_counts l join n_sessions r 
        using(client)
    
)

select * from results

'''

df = query_database(connection, sql)
df.head()
```

```python
for client, df_ in df.groupby('client'):
    df_ = pd.DataFrame(df_)
    df_['percent'] = df_['n_actions'] / df_['n_sessions'] * 100
    df_ = df_[['action', 'percent']]
    df_ = df_.sort_values('percent', ascending=False)
    display(client)
    display(df_)
```
