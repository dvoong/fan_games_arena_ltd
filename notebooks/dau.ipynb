```python
sql = '''
select 
    cast(createdAt as date) as 'date', 
    count(distinct currentUser) as 'dau'
    
from LogPlusPROD 

where createdAt >= '2019-09-01'

group by cast(createdAt as date) 

order by 1
'''
df = query_database('logs', sql)
df.head()
fig, ax = plt.subplots(figsize=(20, 5))
fig.set_title('DAU')
ax.plot(df['date'], df['dau']);
```

```python
sql = '''
select 
    createdAt, 
    currentUser,
    userDevice
    
from LogPlusPROD

'''
df_logs = query_database('logs', sql)
df_logs.head()
```

```python
sql = '''
select id, DeviceOS from FGAUserDevices
'''
df_devices = query_database('users', sql)
df_devices.head()
```

```python
import datetime
import pandas as pd

df = pd.merge(df_logs, df_devices, left_on='userDevice', right_on='id')
df['createdAt'] = pd.to_datetime(df['createdAt'])
df = df[df['createdAt'] < datetime.datetime(2018, 6,1 )]
df = df.drop(['id', 'userDevice'], axis=1)
df['createdAt'] = pd.to_datetime(df['createdAt']).dt.date
df = df.groupby(['createdAt', 'DeviceOS'])[['currentUser']].nunique()
df = df.reset_index()
fig, ax = plt.subplots(figsize=(20, 5))
for client, df_ in df.groupby('DeviceOS'):
    ax.plot(df_['createdAt'], df_['currentUser'], label=client)
    ax.legend()
df.head()
```
